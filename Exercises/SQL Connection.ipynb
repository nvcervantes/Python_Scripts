```python
import os
import pymysql
import pandas as pd

host = os.getenv('localhost')
port = os.getenv('3306')
user = os.getenv('shie')
password = os.getenv('password')
database = os.getenv('clsa')

conn = pymysql.connect(
    host=host,
    port=int(3306),
    user="root",
    passwd=password,
    db="clsa",
    charset='utf8mb4')

df = pd.read_sql_query("SELECT * FROM clsa_incidents",
    conn)
df.tail(10)
```
